```python
# Have to create anndata from our csv file. Oh dear..
# Ok let's start with grabbing the markers and sample details first
markers_info = pd.read_csv("/Users/putri.g/Documents/GitHub/SuperCellCyto-analysis/data/trussart_cytofruv/metadata/Panel.csv")
```

```python
markers_info["marker_name"] = markers_info["fcs_colname"] + "_" + markers_info["antigen"] + "_asinh_cf5"
markers_info.set_index("marker_name", inplace=True)
markers_info.head(10)
```

```python
# separate the expression matrix from other info about the samples
uncorrected_exp_mat = uncorrected_out[markers_info.index.to_numpy()]
uncorrected_exp_mat.head()
```

```python
# extract the supercell details for obs slot
uncorrected_obs = uncorrected_out[['sample_id', 'SuperCellId', 'condition', 'patient_id', 'batch', 'FlowSOM_cluster', 'FlowSOM_metacluster']]
uncorrected_obs.set_index("SuperCellId", inplace=True)
# maybe need to do this so i can subset the string columns properly?
uncorrected_obs = uncorrected_obs.astype('category')
uncorrected_obs.head()
```

```python
uncorrected_adata = ad.AnnData(
    X = uncorrected_exp_mat, 
    var = markers_info,
    obs = uncorrected_obs
)
```

```python
# do the same for cycombine. Maybe i should just merge everything into 1 cell?
# separate the expression matrix from other info about the samples
cycombine_exp_mat = cycombine_out[markers_info.index.to_numpy()]
# extract the supercell details for obs slot
cycombine_obs = cycombine_out[['sample', 'SuperCellId', 'condition', 'patient_id', 'batch', 'FlowSOM_cluster', 'FlowSOM_metacluster']]
cycombine_obs.set_index("SuperCellId", inplace=True)
# maybe need to do this so i can subset the string columns properly?
cycombine_obs = cycombine_obs.astype("category")
cycombine_adata = ad.AnnData(
    X = cycombine_exp_mat, 
    var = markers_info,
    obs = cycombine_obs
)
cycombine_adata
```

```python
# do the same for cytofruv. Maybe i should just merge everything into 1 cell?
# separate the expression matrix from other info about the samples
cytofruv_exp_mat = cytofruv_out[markers_info.index.to_numpy()]
# extract the supercell details for obs slot
cytofruv_obs = cytofruv_out[['sample_id', 'SuperCellId', 'condition', 'patient_id', 'batch', 'FlowSOM_cluster', 'FlowSOM_metacluster']]
cytofruv_obs.set_index("SuperCellId", inplace=True)
# maybe need to do this so i can subset the string columns properly?
cytofruv_obs = cytofruv_obs.astype("category")
cytofruv_adata = ad.AnnData(
    X = cytofruv_exp_mat, 
    var = markers_info,
    obs = cytofruv_obs
)
```

```python
# Need to run PCA to calculate some metrics. Set the n_comps to 20 as the default 50 is way too many
# Also calculate neighbours
# Let's try it on one first. Uncorrected supercells

sc.pp.pca(uncorrected_adata, n_comps=20)
sc.pp.neighbors(uncorrected_adata)

sc.pp.pca(cytofruv_adata, n_comps=20)
sc.pp.neighbors(cytofruv_adata)

sc.pp.pca(cycombine_adata, n_comps=20)
sc.pp.neighbors(cycombine_adata)
```
